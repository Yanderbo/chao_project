# 1.import the libraries

```python
import os
import re
import seaborn as sns
from numpy.fft import *
import pandas as pd
from scipy import signal
from scipy import stats
import matplotlib.pyplot as plt
```

# 2.Read the signal files

Only use the acc files for a demo

```python
folder='Bearing'+idx
name=['h','m','s','0.000001s','hori','verti']
folder_path=r'../input/ieee-phm-2012-data-challenge/ieee-phm-2012-data-challenge-dataset-master/Learning_set'
path=folder_path+'/'+str(folder)
    
file_dir=[]
for file in os.listdir(path):
    if re.match('acc.*',file)!=None:
        file_dir.append(file)
```

*2.5 Choose the params for denoising*

```python
df=pd.read_csv(r'../input/ieee-phm-2012-data-challenge/ieee-phm-2012-data-challenge-dataset-master/Learning_set/Bearing1_1/acc_00001.csv',names=name)
plt.plot(df.hori)

hori=signal.savgol_filter(df.hori,49,3)
plt.plot(hori)
```

# 3.Create the function

```python
def denoise_and_toimg(idx):
            
    for i in range(len(file_dir)):
        df=pd.read_csv(path+'\\'+file_dir[i],names=name)
        hori=signal.savgol_filter(df.hori,49,3)
        hf, ht, hZxx = signal.stft(df.hori,fs=25600,nperseg=6400)
        plt.pcolormesh(ht,hf,np.abs(hZxx))
    
        plt.axis('off')
        plt.savefig('./data/train/'+str(idx)+'/h'+str(file_dir[i])+'.png',bbox_inches='tight',pad_inches=0)       
        verti=signal.savgol_filter(df.verti,49,3)
        vf, vt, vZxx = signal.stft(df.hori,fs=25600,nperseg=6400)
        plt.pcolormesh(vt,vf,np.abs(vZxx))
        #plt.colorbar()
        plt.axis('off')
        #plt.xticks([])
        #plt.yticks([])
        plt.savefig('./data/train/'+str(idx)+'/v'+str(file_dir[i])+'.png',bbox_inch='tight',pad_inches=0)
        
     
```

# 4.take a look
